```python
from bs4 import BeautifulSoup
```

```python
import requests
import re
```

```python
pat = re.compile(r"[a-zA-Z0-9!?.,'\" ]")
pat1 = re.compile(r"[^a-zA-Z0-9!?,.'\" ]")
pat_garbage = re.compile(r"[a-zA-Z]")
```

```python
url_roleplay = ""
url_survey =""
url_in_page=[]
url_introduce = "https://www.google.com/search?q=%EC%98%A4%ED%94%BD+%EC%8A%A4%ED%81%AC%EB%A6%BD%ED%8A%B8&oq=%EC%98%A4%ED%94%BD+%EC%8A%A4%ED%81%AC%EB%A6%BD%ED%8A%B8&aqs=chrome..69i64j5l5.2827j0j7&sourceid=chrome&ie=UTF-8"
def auto_finding():
    pass_url = ['www.gameshot.net','www.jinjja-eng.com','www.gseek.kr','bcmjuan.spworks.co.kr','play.google.com','www.facebook.com','www.instagram.com','books.google.co.kr','www.youtube.com','www.hackers.co.kr','www.pagodastar.com']
    visited = []
    base ="http://www.google.com"
    url = url_introduce
    html = download('get',url)
    dom = BeautifulSoup(html.text,'lxml')
    url_in_page = [_.attrs['href'] for _ in dom.select(".r > a") if requests.compat.urlparse(_.attrs['href'])[1] not in pass_url and 'download' not in _.attrs['href']]
    visited.extend(url_in_page)
    next_page = [_.attrs["href"] for _ in dom.select("#foot  a")]
    np = [requests.compat.urljoin(base,_) for _ in next_page]
    for i in range(len(next_page)):
        url = np[i]
        html = download('get',url)
        dom = BeautifulSoup(html.text,'lxml')
        url_in_page = [_.attrs['href'] for _ in dom.select(".r > a")]
        [visited.append(_) for _ in url_in_page if _ not in visited and requests.compat.urlparse(_)[1] not in pass_url]
    return visited
def auto_script(url):
    html = download('get',url)
    dom = BeautifulSoup(html.text,'lxml')
    script = [_ for _ in [pat1.sub("",_.text) for _ in dom.select('p')] if re.match(pat_garbage,_)]
    return url , script
result = auto_finding()
script_list = [auto_script(item) for item in result]
script_list = [sct for sct in script_list if len(sct[1])>10]
```
